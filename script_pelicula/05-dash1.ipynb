```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
```

```python
pio.renderers.default = "notebook"
```

```python
movies = pd.read_csv(r"../data/process/h_movie.csv")
```

```python
movies.filter(items=["movieid", "title", "rating_promedio", "rating_conteo"]).head(10)
```

```python
movies.info()
```

```python
rating_por_pelicula = movies.groupby("title")["rating_conteo"].sum().sort_values(ascending=False).head(10)
rating_por_pelicula
# rating_por_pelicula = movies.groupby('title').apply(
#     lambda x: x.sort_values('rating_promedio', ascending=False).head(10)
# )
```

```python
rating_por_pelicula.plot(
    kind = "line",
    figsize=(10,4),
    title="Top 10 Pel√≠culas mas vistas ",
    xlabel = "Pel√≠cula",
    ylabel = "N√∫mero de Vistas"
)
plt.grid(False)
plt.xticks(rotation=45, ha="right")
plt.show()
```
