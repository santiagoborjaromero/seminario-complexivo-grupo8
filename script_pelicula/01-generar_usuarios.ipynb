# Generar Usuarios

```python
import pandas as pd
from faker import Faker
import random
```

```python
fake = Faker('es_ES')  # o 'es_ES' 
```

```python
rango = range(1,138494)
```

```python
data = []
userid = 0
for i in rango:
    nombre = fake.name()
    email = nombre.lower().replace(" ", ".") + "@ejemplo.com"
    edad = random.randint(18, 80)
    genero = random.choice(['M', 'F'])
    nogenre = random.choice([1, 0])
    action = random.choice([1, 0])
    adventure  = random.choice([1, 0])
    animation = random.choice([1, 0])
    children = random.choice([1, 0])
    comedy = random.choice([1, 0])
    crime = random.choice([1, 0])
    documentary = random.choice([1, 0])
    drama = random.choice([1, 0])
    fantasy = random.choice([1, 0])
    film_noir = random.choice([1, 0])
    horror = random.choice([1, 0])
    imax = random.choice([1, 0])
    musical   = random.choice([1, 0])
    mystery = random.choice([1, 0])
    romance = random.choice([1, 0])
    sci_fi = random.choice([1, 0])
    thriller = random.choice([1, 0])
    war_western  = random.choice([1, 0])

    provincia = random.choice( [
        "Azuay",
        "Bolívar",
        "Cañar",
        "Carchi",
        "Chimborazo",
        "Cotopaxi",
        "El Oro",
        "Esmeraldas",
        "Galápagos",
        "Guayas",
        "Imbabura",
        "Loja",
        "Los Ríos",
        "Manabí",
        "Morona Santiago",
        "Napo",
        "Orellana",
        "Pastaza",
        "Pichincha",
        "Santa Elena",
        "Santo Domingo de los Tsáchilas",
        "Sucumbíos",
        "Tungurahua",
        "Zamora Chinchipe"
    ])
    
    fila = [i, nombre, email, edad, genero, provincia, nogenre,action,adventure,animation,children,comedy,crime,documentary,drama,fantasy,film_noir,horror,imax,musical,mystery,romance,sci_fi,thriller,war_western]
    data.append(fila)
```

```python
columnas = ['userid', 'nombre', 'email', 'edad', 'genero', 'provincia', 'nogenre','action','adventure','animation','children','comedy','crime','documentary','drama','fantasy','film_noir','horror','imax','musical','mystery','romance','sci_fi','thriller','war_western'] 
```

```python
df = pd.DataFrame(data, columns=columnas)
```
