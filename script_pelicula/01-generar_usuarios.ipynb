```python
df = pd.DataFrame(data, columns=columnas)
```

```python
def sin_tilde(email):
    replacements = (("á", "a"), ("é", "e"), ("í", "i"), ("ó", "o"), ("ú", "u"))
    for a, b in replacements:
        email = email.replace(a, b).replace(a.upper(), b.upper())
    return email
```

```python
df['email'] = df['email'].apply(sin_tilde)
```
