# **Framework DAMA**

Es un  framework que nos permite  trabajar  en el Gobierno y Gestion del Dato, usado mayormente en  America Latina,  a continuación  una  explicación de la  calidad de  dato sus procesos y   sus dimensiones.

En conclusión

DAMA aporta  mucho valor a tener una concientizar a las empresas la calidad de datos  que  tienen, generar planes de acción  que les permita   mejorar su calidad de datos  y por ende tomar  mejores  decisiones  a nivel de  negocio
