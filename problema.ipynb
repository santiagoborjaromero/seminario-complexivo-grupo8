# ANALÍTICA CON PYTHON

### Construcción de un Sistema de Recomendación Híbrido de Películas.

#### Objetivo

Desarrollar un sistema de recomendación híbrido funcional que integre filtrado basado en contenido y filtrado colaborativo. El sistema deberá ser capaz de, para un usuario específico, generar una lista de películas recomendadas que no solo se parezcan a lo que ya le gusta, sino que también sean populares entre otros usuarios con gustos similares.

#### Antecedentes 

Este caso utiliza el dataset "MovieLens 20M". Los sistemas de recomendación más potentes no usan una sola técnica, sino que son híbridos: combinan múltiples estrategias. El reto aquí es construir un sistema que entienda a un usuario desde dos ángulos a la vez:
- ¿Qué le gusta? (Analizando el contenido de las películas que ha valorado, usando géneros y tags).
- ¿A quién se parece? (Analizando sus ratings y comparándolos con los de usuarios similares).

Al combinar ambos enfoques, el sistema puede ofrecer recomendaciones más precisas, personalizadas e incluso sorprendentes.

#### Actividades para el estudiante

Pipeline:

- Descargar y cargar los archivos rating.csv, movie.csv y tag.csv.
- Para cada película, crear un "perfil de contenido" combinando sus géneros (movie.csv) y todas las etiquetas (tags) asociadas a ella (tag.csv). Usar TF-IDF para convertir estos perfiles de texto en vectores numéricos y calcular una matriz de similitud de contenido.
- Utilizar el archivo rating.csv para entrenar un modelo de filtrado colaborativo (por ejemplo, usando la librería Surprise o implementando un algoritmo de factorización de matrices como SVD) que pueda predecir la calificación que un usuario le daría a una película que no ha visto.
- El pipeline debe producir dos artefactos clave: la matriz de similitud de contenido y el modelo colaborativo entrenado.

API:
- Crear una API con un endpoint principal, /recommendations/hybrid/{user_id}.
- La lógica de este endpoint será: primero, usar el modelo colaborativo para obtener una lista inicial de, por ejemplo, 50 películas con la calificación predicha más alta para ese usuario. Luego, re-ordenar (re-rank) esa lista, dándoles un empujón extra en el ranking a aquellas películas que sean temáticamente similares a las que el usuario ya ha calificado con 5 estrellas.

Dashboard:
- Desarrollar un dashboard donde un usuario pueda "iniciar sesión" (seleccionando un user_id).
- El dashboard mostrará una lista de las películas que el usuario ha calificado más alto.
- La funcionalidad principal será la lista de "Top 10 Recomendaciones Híbridas para ti", generada por la API, explicando quizás por qué se recomienda cada película (ej. "Porque te gustó 'Matrix' y a usuarios similares a ti les encantó 'Inception'").
