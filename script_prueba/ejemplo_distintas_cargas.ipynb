```python
sheets = ["Orders", "Returns", "People"]
diccionario = {}
```

```python
for sheet in sheets:
    print(sheet)
```

```python
for sheet in sheets:
    df_temporal = pd.read_excel(r"../data/Sample-Superstore.xls", sheet_name=sheet)
    diccionario[sheet] = df_temporal
```

# Unir data frames

```python
df_pedidos = diccionario["Orders"]
df_devoluciones = diccionario["Returns"]
df_personas = diccionario["People"]
```

```python
df_pedidos_completo = pd.merge(df_pedidos, df_devoluciones, on="Order ID", how="left")
```

```python
df_pedidos_completo.info()
```

```python
df_pedidos_completo = pd.merge(df_pedidos_completo, df_personas, on="Region", how="left")
```
